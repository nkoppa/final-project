```python
DATA_DIR = "/Users/niko2485/Library/CloudStorage/OneDrive-UCB-O365/Desktop/EarthScienceCourse/Final project"
os.makedirs(DATA_DIR, exist_ok=True)

ERA5_DIR = os.path.join(DATA_DIR, "era5land_hourly")
os.makedirs(ERA5_DIR, exist_ok=True)

```

```python
c = cdsapi.Client()

years = list(range(1980, 2024))  
months = ["03", "04", "05", "06"]
days   = [f"{d:02d}" for d in range(1, 32)]
hours  = [f"{h:02d}:00" for h in range(24)]

for year in years:
    out_file = os.path.join(ERA5_DIR, f"era5land_t2m_india_MAMJ_{year}.nc")
    if os.path.exists(out_file):
        print(f"Already have {out_file}, skipping.")
        continue

    print(f"Requesting ERA5-Land t2m for {year}...")
    c.retrieve(
        "reanalysis-era5-land",   # 
        {
            "product_type": "reanalysis",
            "variable": ["2m_temperature"],
            "year": str(year),
            "month": months,
            "day": days,
            "time": hours,
            # N, W, S, E (rough India box)
            "area": [38, 68, 6, 98],
            "format": "netcdf",
        },
        out_file,
    )


```

Note: While downloading the data from CDS, I did not realise that i was downloading zipped files. And since it took almost 3 hours to download all the data, I decided to use chatgpt to write a code to extract the .nc files from the zipped folders. I wrote the code below using chatgpt's help. 

```python
nc_files = sorted(glob.glob(os.path.join(ERA5_DIR, "*.nc"))) 
print("Found", len(nc_files), "wrapper files")

for f in nc_files:
    print("Processing:", os.path.basename(f))
    with open(f, "rb") as fh:
        sig = fh.read(2)

    if sig != b"PK":
        print(" not a zip, skipping")
        continue

    # open as zip and find the actual netcdf inside
    with zipfile.ZipFile(f, "r") as z:
        members = [m for m in z.namelist() if m.endswith(".nc")]
        if len(members) == 0:
            print("  -> no .nc inside, skipping")
            continue

        inner_nc = members[0] 
        # extract inner file to the same directory
        z.extract(inner_nc, ERA5_DIR)

        # build a new name based on the original file (year)
        base = os.path.basename(f)  
        year = base.split("_")[-1].split(".")[0]  
        new_name = f"era5land_t2m_india_MAMJ_{year}_real.nc"
        new_path = os.path.join(ERA5_DIR, new_name)

        # move/rename extracted file
        old_path = os.path.join(ERA5_DIR, inner_nc)
        os.rename(old_path, new_path)
        print("  -> extracted and renamed to:", new_name)

    # delete the original zip-wrapper file to keep folder clean
    os.remove(f)
    print("  -> deleted wrapper:", base)

print("Done fixing files.")
```
